```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
empresas = ['Apple', 'Google', 'Amazon', 'Microsoft', 'Tencent', "McDonald's", 'Visa', 'Facebook', 'Alibaba', 'Louis Vitton']
```

```python
bilhoes = [947.062, 819.573,
          705.646, 611.460,
          214.023, 196.526,
          191.421, 186.421,
          169.966, 124.273]
```

```python
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo >;(')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.show()
```

```python
#redimensionando
plt.rcParams["figure.figsize"] = (15, 8)
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#conhecendo o fontsize
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo >;(', fontsize=20)
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#mudando a cor e legendando 
plt.plot(empresas, bilhoes, color = 'pink')
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilhões', fontsize=15)
plt.title('Empresas mais ricas do mundo >;(', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=10)
```

```python
#aumentar a expessura da linha
plt.plot(empresas, bilhoes, color = 'pink', lw=6)
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilhões', fontsize=15)
plt.title('Empresas mais ricas do mundo >;(', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=10)
```

```python
#adicionando marcador
plt.plot(empresas, bilhoes, color = 'pink', marker = 'o')
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilhões', fontsize=15)
plt.title('Empresas mais ricas do mundo >;(', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=10)
```

```python
#adicionando linha de grade
plt.plot(empresas, bilhoes, color = 'pink', marker = 'o')
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilhões', fontsize=15)
plt.title('Empresas mais ricas do mundo >;(', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=10)
plt.grid(True)
```

```python
#mudando o tipo de gráfico e colocando a legenda acima
plt.bar(empresas, bilhoes, color = 'pink') 
plt.title('Empresas mais ricas do mundo >;(')
plt.xlabel('Empresas', fontsize=15)
plt.ylabel('Bilhões', fontsize=15)
plt.title('Empresas mais ricas do mundo >;(', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=10)
plt.grid()
```

```python
#gráfico simples
plt.bar(empresas, bilhoes, color='purple') 
```

```python
cores = ['black', 'dimgray', 'dimgrey', 'gray', 'grey', 'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey']
```
